```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.utils import shuffle
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.naive_bayes import MultinomialNB
```

```python
def load_data():
    path = "./data/agaricus-lepiota.data"
    df = pd.read_fwf(path, header = None)
    df = df[0].str.split(',',expand = True)
    return df
```

```python
def draw(df):
    for i in range(1, 23, 1):
        print("Feature {}:".format(i))
        print("Total:")
        freq = df[i].value_counts()
        freq.plot.bar(rot = 0)
        plt.show()
        
        print("Label is 'e':")
        df_e = df[(df[0] == 'e')]
        freq_e = df_e[i].value_counts()
        freq_e.plot.bar(rot=0)
        plt.show()
        
        print("Label is 'p':")
        df_p = df[(df[0] == 'p')]
        freq_p = df_p[i].value_counts()
        freq_p.plot.bar(rot=0)
        plt.show()
```

```python
def label_encoding(y):
    labelencoder = LabelEncoder()
    y = labelencoder.fit_transform(y)
    return y
```

```python
def with_Laplace_smoothing(X_train, X_test, y_train, y_test):
    model = MultinomialNB()
    model.fit(X_train, y_train)
    return model.score(X_train, y_train), model.score(X_test, y_test)
```

```python
def without_Laplace_smoothing(X_train, X_test, y_train, y_test):
    model = MultinomialNB(alpha = 0)
    model.fit(X_train, y_train)
    return model.score(X_train, y_train), model.score(X_test, y_test)
```

```python
def holdout_validation(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3) 
    
    with_L_train_score, with_L_test_score = with_Laplace_smoothing(X_train, X_test, y_train, y_test)
    without_L_train_score, without_L_test_score = without_Laplace_smoothing(X_train, X_test, y_train, y_test)
    
    print("Training set score(with Laplace smoothing, holdout validation): {}".format(with_L_train_score))
    print("Testing set score(with Laplace smoothing, holdout validation): {}".format(with_L_test_score))    
    print("Training set score(without Laplace smoothing, holdout validation): {}".format(without_L_train_score))
    print("Testing set score(without Laplace smoothing, holdout validation): {}".format(without_L_test_score))    
```

```python
def k_fold(X, y):
    kf = KFold(n_splits = 3)
    
    total_with_L_train_score, total_with_L_test_score = np.array([]), np.array([])
    total_without_L_train_score, total_without_L_test_score = np.array([]), np.array([])
    
    for train_index, test_index in kf.split(X):
        X_train, X_test, y_train, y_test = X[train_index], X[test_index], y[train_index], y[test_index]
        with_L_train_score, with_L_test_score = with_Laplace_smoothing(X_train, X_test, y_train, y_test)
        without_L_train_score, without_L_test_score = without_Laplace_smoothing(X_train, X_test, y_train, y_test)
        total_with_L_train_score = np.append(total_with_L_train_score, with_L_train_score)
        total_with_L_test_score = np.append(total_with_L_test_score, with_L_test_score)
        total_without_L_train_score = np.append(total_without_L_train_score, without_L_train_score)
        total_without_L_test_score = np.append(total_without_L_test_score, without_L_test_score)
        
    print("Training set average score(with Laplace smoothing, K-fold cross-validation): {}".format(np.mean(total_with_L_train_score)))
    print("Testing set average score(with Laplace smoothing, K-fold cross-validation): {}".format(np.mean(total_with_L_test_score)))    
    print("Training set average score(without Laplace smoothing, K-fold cross-validation): {}".format(np.mean(without_L_train_score)))
    print("Testing set average score(without Laplace smoothing, K-fold cross-validation): {}".format(np.mean(without_L_test_score)))    
```

```python
df = load_data()
#draw(df)
df = shuffle(df) #index也洗了（在data裡一定有至少一筆相同的資料，但index不一定是原本的index）
X = df.drop(columns=[0]) #把feature提取出來
y = df[0] #label 
X = pd.get_dummies(X) #onehot encoding
y = label_encoding(y) #already array type
X = X.to_numpy() #to array type

holdout_validation(X, y)
k_fold(X, y)
```

```python
draw(df)
```
